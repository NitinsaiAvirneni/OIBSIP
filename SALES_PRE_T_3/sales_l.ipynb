```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data=pd.read_csv("Advertising.csv")
```

```python
data.head
```

```python
data.shape
```

```python
data.info()
```

```python
X = data.drop(["Id","Sales"],axis=1)
Y = data['Sales']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Train data vs Test data')
ax1.plot(X_train, Y_train)
ax2.plot(X_test,Y_test)


plt.show()
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
training_data_prediction = model.predict(X_test)
```

```python
print(training_data_prediction)
```

```python
error_score = metrics.r2_score(Y_test, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
