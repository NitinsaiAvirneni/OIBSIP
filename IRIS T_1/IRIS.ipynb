```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('Iris.csv')
```

```python
print(data.head)
```

```python
x1=data["PetalLengthCm"]
y1=data["PetalWidthCm"]
plt.scatter(x1,y1,label="Petals",color="red",marker="H",s=20)
plt.xlabel('Petal Length Cm')
plt.ylabel('Petal Length Cm')
plt.show()
```

```python
x2=data["SepalLengthCm"]
y2=data["SepalWidthCm"]
plt.scatter(x2,y2,label="Petals",color="red",marker="H",s=20)
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Length Cm')
plt.show()
```

```python
x3=data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]

print(x3)



y3=data["Species"]
print(y3)

#################making training and testing data ################
x_1 ,x_2,y_1,y_2=train_test_split(x3,y3,test_size=0.2, random_state=0)
#x_1-train
#X_2-test

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_1, y_1)
```
